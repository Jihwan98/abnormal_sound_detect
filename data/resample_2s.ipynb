```python
import librosa
import soundfile as sf
import glob
```

```python
def resample(data_path, orig_sr, re_sr):
    data, sr = librosa.load(data_path, sr=orig_sr)
    resample_data = librosa.resample(data, sr, re_sr)
    
    return resample_data
```

```python
data_list = glob.glob('실내_data/*.wav')
```

```python
data_list[0].split("/")[1]
```

```python
def split_wav(data, sample_rate, start, end):
    start *= sample_rate
    end *= sample_rate
    return data[start:end]
```

```python
data_list[0].split("/")[1].split(".wav")[0]
```

```python
original_sr = 44100
resample_sr = 16000

for path in data_list:
    name = path.split("/")[1].split(".wav")[0]
    print(f"file : {name}")
    resample_data = resample(path, original_sr, resample_sr)
    print(f'길이 : {resample_data.shape[0] / resample_sr}')
    for i in range(int(resample_data.shape[0] / resample_sr)):
        print(f"{i} 번 째")
        out = split_wav(resample_data, resample_sr, i, i+1)
        sf.write(f'실내_16_1s/{name}_{i}.wav', out, resample_sr)
```
