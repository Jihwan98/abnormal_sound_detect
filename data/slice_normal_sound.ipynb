```python
import librosa
import librosa.display
import soundfile as sf
import os
```

```python
files = os.listdir("normal_16")
```

```python
y, sr = librosa.load(os.path.join("normal_16", files[1]), sr=16000)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,10))
librosa.display.waveplot(y, sr, alpha=0.5)
plt.xlabel("Times (s)")
plt.ylabel("Amplitude")
plt.title("Waveform")
```

```python
sample_rate = 16000
```

```python
def split_wav(data, sample_rate, start, end):
    start *= sample_rate
    end *= sample_rate
    return data[start:end]
```

```python
out = split_wav(y, sample_rate, 0, 1)
```

```python
sf.write("test.wav", out, sample_rate)
```

```python
sample_rate = 16000
for file in files:
    print(f"file : {file}")
    name = file.split('.')[1]
    audio_path = os.path.join("normal_16", file)
    data, sr = librosa.load(audio_path, sr=sample_rate)
    print(f" : {data.shape[0] / sample_rate}")
    for i in range(int(data.shape[0]/sample_rate)):
        print(f"{i}  ")
        out = split_wav(data, sample_rate, i, i+1)
        sf.write(f"after/{name}_{i}.wav", out, sample_rate)
```
