```python
import librosa
import librosa.display
import soundfile as sf
import os
```

```python
files = os.listdir("abnormal_16")
```

```python
y, sr = librosa.load(os.path.join("abnormal_16", files[1]), sr=16000)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,10))
librosa.display.waveplot(y, sr, alpha=0.5)
plt.xlabel("Times (s)")
plt.ylabel("Amplitude")
plt.title("Waveform")
plt.show()
```

```python
def split_wav(data, sample_rate, start, end):
    start *= sample_rate
    end *= sample_rate
    return data[start:end]
```

```python
sample_rate = 16000
for file in files:
    print(f"file : {file}")
    name = file.split('.')[1]
    audio_path = os.path.join("abnormal_16", file)
    data, sr = librosa.load(audio_path, sr=sample_rate)
    print(f" : {data.shape[0] / sample_rate}")
    for i in range(round((data.shape[0]/sample_rate)/2)):
        print(f"{i}  ")
        out = split_wav(data, sample_rate, 2*i, 2*(i+1))
        sf.write(f"abnormal_2s/{name}_{i}.wav", out, sample_rate)
```
